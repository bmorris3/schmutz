```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t, f, e, _ = np.loadtxt('schmutz/data/T1_Spitzer.txt', unpack=True, delimiter=',')
from scipy.signal import medfilt
filtered = medfilt(f, (11,)) 
halfptp = filtered.ptp()/2
print(halfptp)
plt.scatter(t, f, s=1)
plt.scatter(t, filtered, s=1)
plt.axhline(halfptp)
plt.axhline(-halfptp)
```
